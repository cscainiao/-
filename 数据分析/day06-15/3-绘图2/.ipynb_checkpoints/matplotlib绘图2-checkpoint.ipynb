```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 使用x和y的关系来绘制圆
f = lambda x:(1-x**2)**0.5
x = np.linspace(-1,1,100)
y = f(x)
plt.plot(x,y,x,-y)
plt.axis('equal')
```

```python
# axis手动调整坐标界限，建议与figsize一起调整
# 但是要保证figsize的比例和坐标界限的比例相同，不然图像会失调
plt.figure(figsize=(6,10))
radian = np.linspace(0,2*np.pi,100)
x = np.cos(radian)
y = np.sin(radian)
plt.plot(x,y)
# value = plt.axis('equal')
# [xmin, xmax, ymin, ymax]
plt.axis([-3,3,-5,5])
```

```python
plt.figure(figsize=(5,3))
axes = plt.subplot(111)
# 可以使用[xmin, xmax, ymin, ymax]
# axes.axis([-3,3,-5,5])
# 也可以单独对x和y轴界限分别设置
axes.set_xlim([-2,2])
axes.set_ylim([-1,1.5])
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
value = plt.axis('scaled')
```

```python
plt.figure(figsize=(5,3))
plt.plot(x,y)
value = plt.axis('equal')
```

### 绘制图例

```python
data1 = Series(data=[20000,26000,30000,36000,45000,55000,56000,58000,57000],
               index=[2010,2011,2012,2013,2014,2015,2016,2017,2018],name='beijing')
data2 = Series(data=[6000,6500,8000,8000,8500,8800,10000,12000,20000],
               index=[2010,2011,2012,2013,2014,2015,2016,2017,2018],name='chengdu')
data3 = Series(data=[10000,12500,13000,14000,15500,15800,20000,23000,30000],
               index=[2010,2011,2012,2013,2014,2015,2016,2017,2018],name='guangzhou')
```

```python
# 分别设置每一条线的标签，然后再调用plt.legend来显示
plt.plot(data1.index,data1.values,label='beijing')
plt.plot(data2.index,data2.values,label=data2.name)
plt.plot(data3.index,data3.values,label=data3.name)
plt.legend()
```

```python
# 通过向legend传入图例名称列表设置
plt.plot(data1.index,data1.values)
plt.plot(data2.index,data2.values)
plt.plot(data3.index,data3.values)
plt.legend([data1.name,data2.name,data3.name])
```

```python
prices = DataFrame({
    'beijing':data1,
    'chengdu':data2,
    'guangzhou':data3
})
prices
```

```python
# 使用plt绘制dataFrame对象
plt.plot(prices)
plt.legend(prices.columns)
plt.xlabel('years')
plt.ylabel('price')
plt.title('price-years',color='blue',fontsize=20)
```

```python
plt.plot(prices.index,prices.beijing,label='beijing')
plt.plot(prices.index,prices.chengdu,label='chengdu')
plt.plot(prices.index,prices.guangzhou,label='guangzhou')

# loc控制图例的位置
# ncol控制图例的显示列数
plt.legend(loc=[0.1,1],ncol=3)
```

### 线的外观设置

- linestyle(ls) 设置线的风格
- linewidth(lw) 设置线宽
- alpha  透明度
- color  颜色
- marker 点形
- markersize  点大小
- markeredagecolor 点的边界颜色
- markeredagewidth 点的边界宽度
- markerfacecolor  点的主体颜色

#### linestyle 
-\--\-.\:\steps\dashes

```python
# dashes=[线段长度1,空白长度1,线段长度2,空白长度2...]
plt.plot(data1.index,data1.values,dashes=[8,1,2,1])
```

```python
# 简写形式
plt.plot(data1.index,data1.values,ls='steps')
```

### linewidth lw

```python
plt.plot(data1.index,data1.values,lw=3)
```

```python
plt.plot(data1.index,data1.values,alpha=0.4)
```

```python
# color 
# red blue
# #00ff
# [0.4,0.6,0.8]
# plt.plot(data1.index,data1.values,color='#0000ff')
plt.plot(data1.index,data1.values,color=[0.31213,0.6,0.2])
```

### marker

```python
plt.plot(data1.index,data1.values,marker='d',markersize=20)
```

### markeredage

```python
# 绘制圆形
plt.plot(data1.index,data1.values,
         marker='o',
         markersize=30,
         markeredgewidth=1,
         markeredgecolor='orange',
         markerfacecolor='white',
         ls='None')
# 绘制三脚架
plt.plot(data1.index,data1.values,
         marker='2',
         markersize=30,
         ls='None',
         color='orange')
```

```python
# ls支持多参数连用的方法
# 颜色 线形 点形 连接在一起，拼成一个字符串
plt.plot(data1.index,data1.values,'r--o',data2.index,data2.values,'b-.D')
```

```python
# 多条线的统一设置
plt.plot(data1.index,data1.values,data2.index,data2.values,ls='--',color='red')
```
