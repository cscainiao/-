    - eg: lines = plt.plot();   line = lines[0]
    - line.set_linewith()

##### 对坐标系使用一系列的setter方法
- axes = plt.subplot()获取坐标系

```python
# 可以使用列表保存plot函数的返回值【图中的所有对象】
lines = plt.plot(x,x,x,x*2,x,x/2)
lines[0].set_color('red')
lines[1].set_linewidth(5)
lines[2].set_linestyle('--')
```

```python
# 获取子画布
axes = plt.subplot(111)
# 坐标轴界限
axes.set_xlim(-10,10)
axes.set_ylim(-10,10)
# 坐标轴标签
axes.set_xlabel('X_Label',color='r',fontsize=15)
axes.set_ylabel('Y_Label',color='b',fontsize=15,rotation=50)
# 坐标轴刻度、刻度标签
axes.set_xticks([-10,0,10])
axes.set_yticks([-10,-5,0,5,10])
axes.set_xticklabels(['min','0','max'])
axes.set_yticklabels(['lower','low','middle','high','higher'])
# 画布标题
axes.set_title('Axes-Title',color='green',fontsize=30)
# 画布背景色
axes.set_facecolor('yellow')
```

```python
# 使用plt绘制上面的图形
# 坐标轴界限
plt.axis([-10,10,-10,10])
# 坐标轴标签
plt.xlabel('X_Label',color='red',fontsize=15)
plt.ylabel('Y_Label',color='blue',fontsize=15)
# 坐标轴刻度、刻度标签
plt.xticks([-10,0,10],['min','0','max'])
plt.yticks([-10,-5,0,5,10],['lower','low','middle','high','higher'])

# 画布标题
plt.title('Axes-Title')

# 画布背景色
```

## 三、2D图形

### 直方图

【直方图的参数只有一个x！！！不像条形图需要传入x,y】

hist()的参数
+ bins  
可以是一个bin数量的整数值，也可以是表示bin的一个序列。默认值为10
+ normed  
如果值为True，直方图的值将进行归一化处理，形成概率密度，默认值为False
+ color  
指定直方图的颜色。可以是单一颜色值或颜色的序列。如果指定了多个数据集合，颜色序列将会设置为相同的顺序。如果未指定，将会使用一个默认的线条颜色
+ orientation  
通过设置orientation为horizontal创建水平直方图。默认值为vertical

```python
ages = np.random.randint(1,110,size=100)

# horizontal 横向
# vertical   纵向
# rwidth 调整直方图的宽度
plt.hist(ages,bins=5,normed=True,color='blue',orientation='horizontal',rwidth=0.9)
```

## <font color = red>四、图形内的文字、注释、箭头(自学)</font>

控制文字属性的方法:  


| pyplot函数     | API方法                | 描述                     |
| :-------------: |:------------------------------:| :---------------------------------:|
| text()       |  mpl.axes.Axes.text()       | 在Axes对象的任意位置添加文字     |
| xlabel()      |  mpl.axes.Axes.set_xlabel()   | 为X轴添加标签               |
| ylabel()      |  mpl.axes.Axes.set_ylabel()   | 为Y轴添加标签               |
| title()       | mpl.axes.Axes.set_title()   |  为Axes对象添加标题            |
| legend()      | mpl.axes.Axes.legend()      |  为Axes对象添加图例            |
| figtext()     |  mpl.figure.Figure.text()    |  在Figure对象的任意位置添加文字    |
| suptitle()     | mpl.figure.Figure.suptitle() |  为Figure对象添加中心化的标题     |
| annnotate()    |  mpl.axes.Axes.annotate()    |  为Axes对象添加注释（箭头可选）    |  


所有的方法会返回一个matplotlib.text.Text对象

```python
import numpy  as np
import pandas as pd
from pandas import Series,DataFrame
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,7,100)
y = np.sin(x)

axes = plt.subplot(111)
axes.text(x=0,y=0,s='sin(x)=0',color='r',fontsize=20,rotation=60)
axes.plot(x,y)
```

```python
# figure对象就是父画布
figure = plt.figure(figsize=(10,3))
# figure.text相当于在父画布内添加文字
# 参数x,y 分别表示比例值
figure.text(x=0.1,y=0.5,s='figure_text',color='red',fontsize=30)
# figure.suptitle添加父画布的标题
figure.suptitle('Figure_Title',color='orange',fontsize=30)

axes1 = plt.subplot(1,2,1)
axes1.plot(x,x)
axes1.set_title('Axes1-title')

axes2 = plt.subplot(1,2,2)
axes2.plot(x,y)
axes2.text(x=np.pi/2,y=1,s='max',color='red')
axes2.set_title('Axes2-title')
```

### 注释
annotate()
- xy参数设置箭头指示的位置
- xytext参数设置注释文字的位置  
- arrowprops参数以字典的形式设置箭头的样式  
- width参数设置箭头长方形部分的宽度
- headlength参数设置箭头尖端的长度，  
- headwidth参数设置箭头尖端底部的宽度
- shrink参数设置箭头顶点、尾部与指示点、注释文字的距离（比例值），可以理解为控制箭头的长度
![](15.png)

    如下都是arrowstyle可以选择的风格样式
    
    ``'->'``       head_length=0.4,head_width=0.2
    ``'-['``       widthB=1.0,lengthB=0.2,angleB=None
    ``'|-|'``      widthA=1.0,widthB=1.0
    ``'-|>'``      head_length=0.4,head_width=0.2
    ``'<-'``       head_length=0.4,head_width=0.2
    ``'<->'``      head_length=0.4,head_width=0.2
    ``'<|-'``      head_length=0.4,head_width=0.2
    ``'<|-|>'``    head_length=0.4,head_width=0.2
    ``'fancy'``    head_length=0.4,head_width=0.4,tail_width=0.4
    ``'simple'``   head_length=0.5,head_width=0.5,tail_width=0.2
    ``'wedge'``    tail_width=0.3,shrink_factor=0.5


```python
index = np.arange(0,14,step=1)
data = np.random.randint(10,15,size=14)
data[5] = 32

axes = plt.subplot(111)
axes.plot(index,data)
axes.set_ylim([8,35])

# 添加注释信息
# s 是要添加的文字
# xy是一个列表，来描述注释指向的点
# xytext是一个列表，描述注释文字摆放的位置

# arrowprops字典的键
# width参数设置箭头长方形部分的宽度
# headlength参数设置箭头尖端的长度，
# headwidth参数设置箭头尖端底部的宽度
# shrink参数设置箭头顶点、尾部与指示点、注释文字的距离（比例值），可以理解为控制箭头的长度
axes.annotate(s='annotation',xy=[5,32],xytext=[7,28],arrowprops={
    'width':10,
    'headlength':8,
    'headwidth':15,
    'shrink':0.1
})
```

```python
axes = plt.subplot(111)
axes.plot(index,data)
axes.set_ylim([8,35])
# 第二种箭头的设置方式，使用预定义箭头类型
axes.annotate(s='annotation',xy=[5,32],xytext=[7,28],arrowprops={
    'arrowstyle':'<|-|>'
})
```

绘制3d图形

添加colorbar

```python
from mpl_toolkits.mplot3d.axes3d import Axes3D
plt.figure(figsize=(10,5))
axes = plt.subplot(projection='3d')

x = np.linspace(0,2*np.pi,50)
y = np.linspace(0,2*np.pi,50)

X,Y = np.meshgrid(x,y)

def creat_Z(X,Y):
    return np.cos(X)*0.3 + np.sin(Y)*0.5 + np.sin(X)*np.cos(Y)

Z = creat_Z(X,Y)
p = axes.plot_surface(X,Y,Z,color='red',alpha=0.6,cmap='spring',rstride=2,cstride=2)
plt.colorbar(p,shrink=0.8)
axes.set_xlabel('X-label')
axes.set_ylabel('Y-label')
axes.set_zlabel('Z-label')
axes.set_title('3D-AXES')
```
