```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame

from PIL import Image
from numpy.fft import fft,ifft
```

```python
# 加载图片
image = Image.open('cat.jpg')
```

```python
# 把image对象转换成数据
image_data = np.frombuffer(image.tobytes(),dtype=np.int8)
image_data
```

```python
# i*i = -1
# 傅里叶变换 时域->频域
fft_data = fft(image_data)
fft_data
```

```python
# 滤波操作（把某个值以下的波全部清0）
fft_data[np.abs(fft_data) < 4e5] = 0
```

```python
# 傅里叶变换 频域->时域
ifft_data = ifft(fft_data)
ifft_data
```

```python
# 获取转换后的实数部分并且转成Int8类型
result_data = np.int8(np.real(ifft_data))
```

```python
# 把数据转换成图片
Image.frombytes(data=result_data,mode=image.mode,size=image.size)
```
