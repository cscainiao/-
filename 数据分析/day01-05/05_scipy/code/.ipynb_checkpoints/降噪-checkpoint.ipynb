```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
moon = plt.imread('./moonlanding.png')
moon.shape
```

```python
from numpy.fft import fft,ifft
```

```python
moon_fft = fft(moon) #将描述值的值变成描述变化的值
moon_fft
```

```python
np.abs(moon_fft).mean() #正常的变化范围
```

```python
# 将大于阈值10 波置为0
cond = np.abs(moon_fft) > 10
moon_fft[cond] = 0
```

```python
moon_ifft = ifft(moon_fft) #清除变化太大的值之后 再把描述变化的值变成描述值的值
```
